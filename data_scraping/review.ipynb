```python
import json
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

# Path to your cleaned JSON file
json_path = Path('/Users/mayankjangid/AIMS_Research/data_scraping/recipes_output/recipes_data_clean.json')

# Load recipes
with open(json_path, 'r', encoding='utf-8') as f:
    recipes = json.load(f)

# Display in notebook
for recipe in recipes:
    real_title = recipe.get('title', 'Unknown')
    noisy_title = recipe.get('noisy_title', 'Unknown')
    img_path = "/Users/mayankjangid/AIMS_Research/data_scraping/" + recipe.get('main_image', '')


    # Open and plot
    img = Image.open(img_path)
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Real: {real_title}\nNoisy: {noisy_title}", fontsize=10)
    plt.show()
```
